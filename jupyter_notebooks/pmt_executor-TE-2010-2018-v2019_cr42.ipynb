### Prueba de papermill

```python
test_str = "pmt-TE-2010-2018-v2019_r0v3_cmeteo_cboot_cr42"
```

```python
import numpy as np
import papermill as pm

# Definir la lista de parámetros a explorar
percs_list = [0.97,0]
meteo_vars_list = [1,0]
stations_list = ["UIZ", "AJU", "ATI", "CUA", "SFE", "SAG", "CUT", "PED", "TAH", "GAM", "IZT", "CCA", "HGM", "LPR",
                "MGH", "CAM", "FAC", "TLA", "MER", "XAL", "LLA", "TLI", "UAX", "BJU", "MPA",
                "MON", "NEZ", "INN", "AJM", "VIF"]

num_of_repetitions_list = list(range(3))

# Generar el grid de parámetros
params_grid = [(perc_size, meteo_vars, cur_station, num_test)
               for perc_size in percs_list
               for meteo_vars in meteo_vars_list
               for cur_station in stations_list
               for num_test in num_of_repetitions_list]

# Imprimiendo parametros
for i, (perc_size, meteo_vars,cur_station, num_test) in enumerate(params_grid):
    params = {"perc_size": perc_size, "meteo_vars": meteo_vars, "cur_station": cur_station, 
              "test_identifier": f'{cur_station}_{((str(perc_size))).replace(".", "_")}_Meteo{bool(meteo_vars)}_{num_test}_{i}', "test_str": test_str}
    print(i, params, '\n')
```

```python
# Ejecutar el notebook anterior con cada conjunto de parámetros
for i, (perc_size, meteo_vars,cur_station, num_test) in enumerate(params_grid):
    params = {"perc_size": perc_size, "meteo_vars": meteo_vars, "cur_station": cur_station, 
              "test_identifier": f'{cur_station}_{((str(perc_size))).replace(".", "_")}_Meteo{bool(meteo_vars)}_{num_test}_{i}', "test_str": test_str}
    print(i, params, '\n')
    try:
        if i >-1 : #recuperando ultima corrida
            pm.execute_notebook(
                f"{test_str}.ipynb",  # Nombre del notebook anterior
                f"{test_str}_{i}.ipynb",  # Nombre del archivo de salida
                parameters=params  # Parámetros para la ejecución actual
            )        
    except Exception as e:
        print(f"Ocurrió una excepción al procesar el elemento {i} {params}: {e}")
        continue
```

```python
# Definir los pesos de las diferentes métricas (por ejemplo, para ponderar más la precisión que la MAE)
pesos = {'rmse': 0.5, 'mae': 0.3, 'mape': 0.2}

# Crear una lista con los resultados de los diferentes modelos (en este ejemplo, los resultados están almacenados en un diccionario llamado "resultados")
resultados = dict_valores

# Calcular la puntuación para cada modelo
puntuaciones = {}
for modelo, metricas in resultados.items():
    puntaje = 0
    for metrica, valor in metricas.items():
        if metrica in pesos:
            puntaje += pesos[metrica] * (1 / valor) # Utilizamos el inverso para que los valores más altos correspondan a un mejor desempeño
    puntuaciones[modelo] = puntaje

# Imprimir la puntuación para cada modelo
for modelo, puntaje in puntuaciones.items():
    print(f'El modelo {modelo} tiene una puntuación de {puntaje}')
```

```python
# Encontrar el modelo con la mejor puntuación
mejor_modelo = max(puntuaciones, key=puntuaciones.get)
mejor_puntuacion = puntuaciones[mejor_modelo]

# Imprimir el modelo con la mejor puntuación
print(f'El modelo {mejor_modelo} tiene la mejor puntuación de {mejor_puntuacion}')
```
